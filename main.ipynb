```python
from src.simulation.planner_mode import PlannerInterviewSimulation
from src.simulation.baseline import InterviewSimulation
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
profile_path = "prompts/roles/elder_profile_example.json"
max_turns = 60
planner_instruction_path = "docs\planner-instruction.yaml"
model_type=os.getenv("MODEL_TYPE")
model_base_url=os.getenv("MODEL_BASE_URL")
api_key=os.getenv("API_KEY")
```

```python
def planner_mode_simulate():
    simulation = PlannerInterviewSimulation(
        interviewee_profile_path=profile_path,
        max_turns=(max_turns)*2+1, 
        planner_instruction_path=planner_instruction_path,
        model_type=model_type,
        model_base_url=model_base_url,
        api_key=api_key
    )
    
    simulation.initialize_agents()
    output_file = simulation.run_interview()
    
    summary = simulation.get_summary()
    print("\n" + "="*70)
    print("【访谈摘要】")
    print(f"会话ID: {summary['session_id']}")
    print(f"模式: {summary['mode'].upper()}")
    print(f"总问题数: {summary['total_turns']}")
    print(f"Planner 决策数: {summary['planner_decisions_count']}")
    print(f"结果文件: {output_file}")
    print("="*70)
    
    return output_file
```

```python
def baseline_mode_simulate():
    simulation = InterviewSimulation(
        interviewee_profile_path=profile_path,
        max_turns=max_turns,
        model_type=model_type,
        model_base_url=model_base_url,
        api_key=api_key
    )
    
    simulation.initialize_agents()
    output_file = simulation.run_interview()
    
    # 打印摘要
    summary = simulation.get_summary()
    print("\n" + "="*60)
    print("【访谈摘要】")
    print(f"会话ID: {summary['session_id']}")
    print(f"总问题数: {summary['total_turns']}")
    print(f"结果文件: {output_file}")
    print("="*60)
    
    return output_file
```

```python
# baseline 模式和 planner 模式的仿真
baseline_output_file = baseline_mode_simulate()
```

```python
planner_output_file = planner_mode_simulate()
```
